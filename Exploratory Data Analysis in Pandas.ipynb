# Exploratory Data Analysis in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\alexj\Downloads\Pandas Tutorial\world_population.csv")

df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)    

#float is decimal numbers, utilizing (lambda x: '%.2f' % x) adjusts the decimal points displayed in this case two decimal points
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()        #to find how many null values in the data set by column
```

```python
df.nunique()   #shows unique values of each column
```

```python
df.sort_values(by = "2022 Population", ascending = False).head(10)    #utilize ascending = False to display largest 2022 pop.
```

```python
df.sort_values(by = "World Population Percentage", ascending = False).head(10) 
```

```python
df.corr()                    #to find correlation between columns
```

```python
sns.heatmap(df.corr(), annot = True)

plt.rcParams['figure.figsize'] = (15,5)

plt.show()
```

```python
df.groupby('Continent').mean()
```

```python
df.groupby('Continent').mean().sort_values(by="2022 Population", ascending=False)
```

```python
df2 = df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by="2022 Population", ascending=False)

df2
```

```python
df.columns
```

```python
df3 = df2.transpose()              #use .transpose() to display data differently
df3
```

```python
df3.plot()
```
