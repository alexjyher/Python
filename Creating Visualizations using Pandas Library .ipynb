# Creating Visualizations using Pandas Library 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\alexj\Downloads\Pandas Tutorial\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-muted')
```

```python
df.plot()  #default plot is a line plot
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)    #use subplots to separate the visualizations
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores') 

#use lower case title to add a title, use xlabel for horizontal label, use ylabel for vertical label
```

```python
df.plot(kind = 'bar')     #use bar to get a bar graph
```

```python
df.plot(kind = 'bar', stacked = True)    #use stacked = True to get a stacked bar graph
```

```python
df['Flavor Rating'].plot(kind ='bar')     #add bracket after df to display specific columns
```

```python
df.plot.barh()              #.barh() allows you to display a horizontal bar graph
```

```python
df.plot.barh(stacked = True)  
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')       

#use.scatter and specify x and y values to create a scatter plot
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100)      #use s to change the size of dots 
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', c = 'black')  #use c to change the color of the dot
```

```python
df.plot.hist()   #.hist to display histograms
```

```python
df.plot.hist(bins = 20)             #use bins to control the distribution of variables
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10,5)) #use figsize to increase visual
```

```python
df.plot.pie(y='Flavor Rating')  
```

```python
df.plot.pie(y='Flavor Rating', figsize = (10,5)) 
```
